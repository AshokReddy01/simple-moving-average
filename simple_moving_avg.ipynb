```python
import threading
from time import sleep
import numpy as np
```

```python
# function generate the new random value 
def generateValue():

    v1=np.random.randint(10)
    return v1
```

```python
# create an virual environment like stock moving average we genrate a random value calling a function and the avg them
values=[]
avg=0
period=int(input("Enter the period time : "))
for i in range(0,100):

    v=generateValue() # calling the genratevalue function 
    values.append(int(v)) 
    print("\nvalue generated is {}".format(v))
    print("added to the list")
    
    if len(values)>period:
        values=values[1:]
        avg=sum(values)/period
        print("\nSMA is {}".format(avg))
    else:
        avg=sum(values)/len(values)
        print("\nSMA is {}".format(avg))
```
